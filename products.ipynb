```python
ingredients_list = []
for i in df["ingredients"]:
    # we can't quite do:  big_ingred_list.extend(ingred_list.split(", "))
    # because there are commas within ingredients i.e.  "LIQUID SUGAR (SUGAR, WATER)" is 1 ingredient
    start = 0
    inside = False
    for x,char in enumerate(i):
        if char == "(":
            inside = True
        if char == ")":
            inside = False
        if not inside and char == ",":
            ingredients_list.append(i[start:x].lstrip())
            start = x+1

ingr = Counter(ingredients_list)
most_common = ingr.most_common(17)
most_common 
```

```python
least_common = ingr.most_common()[-20:]
least_common
```

```python
top_flavor_ingreds = count_over.sort_values("rating", ascending=False)["ingredients"].head(10)
big_ingred_list2 = []
for i in top_flavor_ingreds:
    start = 0
    inside = False
    for x,char in enumerate(i):
        if char == "(":
            inside = True
        if char == ")":
            inside = False
        if not inside and char == ",":
            big_ingred_list2.append(i[start:x].lstrip())
            start = x+1

ct2 = Counter(big_ingred_list2)
[(ingred,count) for ingred,count in ct2.most_common() if ingred not in [i for i,x in most_common]]
```
